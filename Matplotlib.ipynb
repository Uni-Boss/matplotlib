# Beginners Guide To Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)
```

```python
print('The first ten x values are: ' + str(x[:10]))
print('The first ten y values are: ' + str(y[:10]))
```

```python
plt.plot(x,y)
plt.show()
```

If you want another series on the same plot, you simply use plt.plot() again before the call to plt.show()

```python
# create another series to plot
z = 1 + np.cos(2 * np.pi * x)

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

# Formatting Your Plots

### Axes
It is standard practice to place the axes in a variable called 'ax'. The below code:

Sets the x and y axes limits
Sets the x and y tick labels

```python
ax = plt.axes()

# set the x-axis limits to -1 and 2.5
ax.set_xlim([-0.3,2.1])
# set the y-axis limits to -1 and 2.5
ax.set_ylim([-0.3,2.1])
# set the x tick labels
ax.set_xticks([0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.50, 1.75, 2.0])
# set the y tick labels
ax.set_yticks([0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.50, 1.75, 2.0])

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

### Grids
You can add a grid by calling ax.grid()

```python
ax = plt.axes()
# place a grid on the plot
ax.grid()
# plot the series
plt.plot(x,y)
plt.plot(x,z)
# show the plot
plt.show()
```

### Line Styles
You can change the appeareance of each series's linestyle by adding an additional agrument in the call to plt.plot().

'-'  is a solid line
'--' is a dashed line
'-.' is a dash-dot line
':'  is a dotted line

There are more stlyes available where you can control the exact pattern and spacing and if you want to research this the matplotlib documentation at https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/linestyles.html describes this further.

```python
ax = plt.axes()
# plot the series
plt.plot(x,y, '-')
plt.plot(x,z, '-.')
# show the plot
plt.show()
```

### Line Colours
You can set the line colours to a host of standard preset colours using another argument in the plt.plot() call.

    'b' : blue
    'g' : green
    'r' : red
    'c' : cyan
    'm' : magenta
    'y' : yellow
    'k' : black
    'w' : white
    
You can also designate a colour by its hex code.

```python
ax = plt.axes()
# plot the series
plt.plot(x,y, c='r')
plt.plot(x,z, c='#7d2ea5')
# show the plot
plt.show()
```

### Markers
To help further differentiate your plots you can use a different marker for each series:

    '.' : point
    ',' : pixel
    'o' : circle
    'v' : triangle 1 down
    '^' : triangle 1 up
    '<' : triangle 1 left
    '>' : triangle 1 right
    '1' : triangle 2 down
    '2' : triangle 2 up
    '3' : triangle 2 left
    '4' : triangle 2 right
    's' : square
    'p' : pentagon
    '*' : star
    'h' : Haxagon stlye 1
    'H' : Haxagon stlye 2
    '+' : plus
    'x' : X
    'D' : diamond
    'd' : thin diamond
    '|' : vertical line
    '_' : horizontal line

```python
ax = plt.axes()
# plot the series
plt.plot(x,y, '.', c='r')
plt.plot(x,z, 'v', c='#7d2ea5')
# show the plot
plt.show()
```

### Annotating Your Plots

As nice as your plot can become, every plot you make should have labels for the axes, a title, and a legend. Without these your reader will have a hard time trying to understand what exactly they are looking at. You can also annotate the plot by providing the coordinates of where the annotation should appear.

```python
ax = plt.axes()

# add a title
plt.title('Sine vs Cosine')

# add the x-axis label
plt.xlabel('time (s)')
# add the x-axis label
plt.ylabel('Position (cm)')

# annotate the plot
plt.annotate(xy=[0.5,1.75], s='Look here')

# plot the series
plt.plot(x,y, '.', c='r')
plt.plot(x,z, 'v', c='#7d2ea5')

# add the legend
plt.legend(['Sine series', 'Cosine series'])

# show the plot
plt.show()
```

```python
ax = plt.axes()

# add a title
plt.title('Sine vs Cosine', fontsize=30)

# add the x-axis label
plt.xlabel('time (s)', fontsize=20)
# add the x-axis label
plt.ylabel('Position (cm)', fontsize=20)

# annotate the plot
plt.annotate(xy=[0.5,1.75], s='Look here', fontsize=10)

# plot the series
plt.plot(x,y, '.', c='r')
plt.plot(x,z, 'v', c='#7d2ea5')

# add the legend
plt.legend(['Sine series', 'Cosine series'], loc='lower left')

# show the plot
plt.show()
```

```python
ax = plt.axes()

# add a title
plt.title('Sine vs Cosine', fontsize=30)

# add the x-axis label
plt.xlabel('time (s)', fontsize=20)
# add the x-axis label
plt.ylabel('Position (cm)', fontsize=20)

# set the x-tick properties
plt.xticks(fontsize=12, rotation=90)
# set the y-tick properties
plt.yticks(fontsize=12, rotation=60)

# annotate the plot
plt.annotate(xy=[0.5,1.75], s='Look here', fontsize=10)

# plot the series
plt.plot(x,y, '.', c='r')
plt.plot(x,z, 'v', c='#7d2ea5')

# add the legend
plt.legend(['Sine series', 'Cosine series'], loc='upper right')

# show the plot
plt.show()
```

### Subplots

Sometimes you might want to display plots next to each other and to do this you use the subplot feature:

plt.subplot( rows, columns, position of plot)

Your rows and columns form a grid and you then position each plot in a segment of that grid:

```python
fig = plt.figure()

plt.subplot(2, 2, 1)
plt.title('Plot 1')
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.title('Plot 2')
plt.plot(x, y)

plt.subplot(2, 2, 3)
plt.title('Plot 3')
plt.plot(x, y)

plt.subplot(2, 2, 4)
plt.title('Plot 4')
plt.plot(x, y)

plt.show()
```

```python
fig = plt.figure()
    
plt.subplot(2, 2, 1)
plt.title('Plot 1')
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.title('Plot 2')
plt.plot(x, y)

plt.subplot(2, 2, 3)
plt.title('Plot 3')
plt.plot(x, y)

plt.subplot(2, 2, 4)
plt.title('Plot 4')
plt.plot(x, y)

plt.subplots_adjust(hspace=0.55)

    
plt.show()

```

Below are some common charts that you will probably want to visualise your data.

### Scatterplot
Scatter plots are great at showing clusters of data and reveal underlying patterns:

```python
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 1000).T

```

```python
plt.scatter(x, y, s=[50], marker='.', c='m')
```

### Histogram
Histograms categorise data by taking their minimum and maximum values and dividing the range into segments called bins. Each datum is placed into a bin and the resulting bin count is displayed.

```python
# plot the y data with one bin
plt.hist(y,1)
plt.show()
```

```python
# plot the y data with 25 bins
plt.hist(y,25)
plt.show()
```

### Box plots
Box plots show a number of statistics about the data through their quartiles (points which divide the data into four equal parts) and their whiskers which indicate the upper and lower quartiles.

```python
plt.boxplot(x)
```

```python
plt.boxplot(x, sym='gx', widths=0.75, notch=True)
```

### Pie Charts
Pie charts show the proportion of each data category when viewed as a whole, and provide a quick and simple comparison.

```python
# create the data values
values = [1,2,3,4,5]

plt.pie(values)
plt.show()
```

```python
# create the data values
values  = [1,2,3,4,5]
colours = ['y', 'g', 'c', '#4286f4', '#a169e5']
labels  = ['A', 'B', 'C', 'D', 'E']
explode = (0, 0.3, 0, 0, 0)

plt.pie(values, colors = colours, labels = labels, explode=explode)
plt.show()
```

### Bar Chart
Like pie charts, bar charts allow you to quickly and easily compare the data:

```python
# create the data values

# values to be shown on the y-axis
values  = [1, 6, 3, 4, 3]
# width of each bar
widths = [0.7, 0.3, 0.2, 0.6, 0.5]
# colour of each bar
colours = ['y', 'g', 'c', '#4286f4', '#a169e5']
# label of  each data point
labels  = ['A', 'B', 'C', 'D', 'E']

plt.bar(labels, values, width=widths, color = colours, align='center')
plt.show()
```
